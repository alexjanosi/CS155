```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
train_file = 'data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

# Lasso Regression

```python
# set alpha values
alph = np.linspace(0.01, 3, 30) 

w = [[],[],[],[],[],[],[],[],[]]

# create data for weight values
for l in alph:
    lr = Lasso(alpha = l)
    lr.fit(X_train, y_train)
    for i in range(9):
        w[i].append(lr.coef_[i])

# plot figures
plt.figure()
for q in range(9):
    plt.plot(alph, w[q])
plt.legend(['w1', 'w2', 'w3', 'w4', 'w5', 'w6', 'w7', 'w8', 'w9'], loc = 1)
plt.xlabel('Alpha')
plt.ylabel('Weight Value')
plt.title('Lasso Regression Regularization')
plt.show()
```

# Ridge Regression

```python
# set alpha values
alph = []
for i in range(1,10001):
    alph.append(i)

w = [[],[],[],[],[],[],[],[],[]]

# create data for weight values
for l in alph:
    ridge = Ridge(alpha = l)
    ridge.fit(X_train, y_train)
    for i in range(9):
        w[i].append(ridge.coef_[i])

# plot figures
plt.figure()
for q in range(9):
    plt.plot(alph, w[q])
plt.legend(['w1', 'w2', 'w3', 'w4', 'w5', 'w6', 'w7', 'w8', 'w9'], loc = 1)
plt.xlabel('Alpha')
plt.ylabel('Weight Value')
plt.title('Ridge Regression Regularization')
plt.show()
    
```
